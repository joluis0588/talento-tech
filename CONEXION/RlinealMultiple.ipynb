```python
# 2. Calcular la transpuesta de X (X')
X_t = X.T
print("--- Transpuesta de X (X') ---")
print(X_t)
print("\n" + "="*30 + "\n")

# 3. Multiplicar la transpuesta de X por X (X'X)
XtX = X_t @ X  # El operador @ significa multiplicaci√≥n de matrices
print("--- Producto de X' por X (X'X) ---")
print(XtX)
print("\n" + "="*30 + "\n")


# 4. Calcular la inversa de (X'X)
XtX_inv = np.linalg.inv(XtX)
print("--- Inversa de (X'X) ---")
print(XtX_inv)
print("\n" + "="*30 + "\n")

# 5. Multiplicar la transpuesta de X por Y (X'Y)
XtY = X_t @ Y
print("--- Producto de X' por Y (X'Y) ---")
print(XtY)
print("\n" + "="*30 + "\n")

# 6. Finalmente, calcular los coeficientes Beta (Œ≤)
# Œ≤ = (X'X)^-1 * (X'Y)
beta_coeficientes = XtX_inv @ XtY
print("--- Coeficientes de Regresi√≥n (Œ≤) ---")
print(beta_coeficientes)

# Mostrar los coeficientes de una forma m√°s legible
print("\n--- Resultados Finales ---")
print(f"Intercepto (Œ≤‚ÇÄ): {beta_coeficientes[0][0]:.4f}")
print(f"Coeficiente X‚ÇÅ (Œ≤‚ÇÅ): {beta_coeficientes[1][0]:.4f}")
print(f"Coeficiente X‚ÇÇ (Œ≤‚ÇÇ): {beta_coeficientes[2][0]:.4f}")
```

```python
# error cuadratico medio
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

# Re-create the dataframe df
data = {'Y': [78.5, 74.3, 104.3, 87.6, 95.9, 109.2, 102.7, 72.5, 93.1, 115.9, 83.8, 113.3, 109.4],
        'X1': [7, 1, 11, 11, 7, 11, 3, 1, 2, 21, 1, 11, 10],
        'X2': [26, 29, 56, 31, 52, 55, 71, 31, 54, 47, 40, 66, 68],
        'X3': [6, 15, 8, 8, 6, 9, 17, 22, 18, 4, 23, 9, 8],
        'X4': [60, 52, 20, 47, 33, 22, 6, 44, 22, 26, 34, 12, 12]}
df = pd.DataFrame(data)

# Crear la matriz X a partir de las variables independientes y agregar una columna de unos para el intercepto
X = df[['X1', 'X2', 'X3', 'X4']].values
X = np.insert(X, 0, 1, axis=1)

# Crear el vector Y a partir de la variable dependiente
Y = df['Y'].values.reshape(-1, 1) # Asegurarse de que Y sea un vector columna

# 2. Calcular la transpuesta de X (X')
X_t = X.T

# 3. Multiplicar la transpuesta de X por X (X'X)
XtX = X_t @ X  # El operador @ significa multiplicaci√≥n de matrices

# 4. Calcular la inversa de (X'X)
XtX_inv = np.linalg.inv(XtX)

# 5. Multiplicar la transpuesta de X por Y (X'Y)
XtY = X_t @ Y

# 6. Finalmente, calcular los coeficientes Beta (Œ≤)
# Œ≤ = (X'X)^-1 * (X'Y)
beta_coeficientes = XtX_inv @ XtY

# Re-create X and Y from the dataframe and calculate Y_hat
X = df[['X1', 'X2', 'X3', 'X4']].values
X = np.insert(X, 0, 1, axis=1)
Y = df['Y'].values.reshape(-1, 1)
Y_hat = X @ beta_coeficientes

mse = mean_squared_error(Y, Y_hat)
print("Mean Squared Error (MSE):", mse)

## error cuadratico medio raiz
rmse = np.sqrt(mse)
print(f"Root Mean Squared Error (RMSE): {rmse}")

### error adsoluto medio
mae = mean_absolute_error(Y, Y_hat)
print(f"Mean Absolute Error (MAE) es: {mae}")

# Calcular error porcentual absoluto medio (MAPE)
# Calculate error MAPE
error = Y - Y_hat
mape = (np.abs(error) / np.abs(Y)).mean() * 100
print(f"Mean Absolute Percentage Error (MAPE): {mape}")

## error cuadratico
media_y = df['Y'].mean()
print(f"La media de y es: {media_y}")
```
