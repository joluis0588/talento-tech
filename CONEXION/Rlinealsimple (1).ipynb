```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de la imagen
x = np.array([1, 2, 2, 3, 4, 4, 5, 6])
y = np.array([2, 3, 4, 4, 4, 6, 5, 7])

# Calcular la regresión lineal
# polyfit(x, y, grado del polinomio) retorna [pendiente, intercepto]
m, b = np.polyfit(x, y, 1)

# Generar los valores predichos de y
y_predicha = m * x + b

# Graficar
plt.figure(figsize=(8, 6))
plt.scatter(x, y, marker='x', color='red', label='Puntos de datos')
plt.plot(x, y_predicha, color='blue', label=f'Línea de regresión: y = {m:.4f}x + {b:.4f}')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal en Python')
plt.grid(True)
plt.legend()
plt.show()

print(f"Pendiente (m): {m:.4f}")
print(f"Intercepto (b): {b:.4f}")
```

```python
# Cálculo de Errores (manualmente)
import numpy as np
import pandas as pd

x = [1,2,2,3,4,4,5,6]
y = [2,3,4,4,4,6,5,7]

d = {"x":x, "y":y}
df = pd.DataFrame(d)

df2=df.copy()
df2["y"]=np.ones(len(df))

X=np.array(df2)
y = np.array(df[["y"]])
betas = np.linalg.pinv(X) @ y
df["ypred"] = X @ betas
df
```

```python
# error = y - ypred
df["y-ypred"] =  df["y"] - df["ypred"]
df
```

```python
# error = y - ypred
df["error^2"] = df["y-ypred"]**2
df
```

```python
error_cuad_total = df["error^2"].sum()
error_cuad_total
```

```python
# MSE = error_cuad_medio
mse = error_cuad_total / len(df)
mse
```

```python
rmse = mse**(1/2)
rmse
```

```python
# Calcular el Error Absoluto Medio (MAE)
mae = df['y-ypred'].abs().sum() / len(df)
mae
```

```python
# Calcular error porcentual absoluto medio (MAPE)
mape = (df['y-ypred'].abs() / df['y'].abs()).mean() * 100
mape
```

```python
## error cuadratico
media_y = df['y'].mean()
print(f"La media de y es: {media_y}")
```

```python
sst = ((df['y'] - media_y)**2).sum()
print(f"El valor de SST es: {sst}")
```

```python
sse = df["error^2"].sum()
r2 = 1 - (sse / sst)
print(f"El valor de R² es: {r2}")
```

```python
from sklearn.metrics import r2_score
print(f"El valor de R² es: {r2_score(df['y'], df['ypred'])}")
```

```python
#problema de regrecion lineal simple
#Una empresa requiere saber si existe una relación confiable entre las inversiones en publicidad y las ventas que se obtienen.

# Variable para la Inversión en Publicidad
inversion = [5000, 5000, 4350, 7900, 6800, 5400, 6900, 3900, 4200, 5780]

# Variable para las Ventas
ventas = [160000, 189380, 139200, 260700, 217600, 183600, 234600, 136500, 138600, 202300]
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# 1. Datos del problema
inversion = np.array([5000, 5000, 4350, 7900, 6800, 5400, 6900, 3900, 4200, 5780])
ventas = np.array([160000, 189380, 139200, 260700, 217600, 183600, 234600, 136500, 138600, 202300])
print(inversion)
print(ventas)
```

```python
# 2. Scikit-learn requiere que la variable X (inversion) sea un array 2D
# El .reshape(-1, 1) lo ajusta a este formato.
X = inversion.reshape(-1, 1)
y = ventas
print(X)
print(y)
```

```python
# 3. Crear y entrenar el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(X, y)
```

```python
# 4. Obtener el intercepto y el coeficiente
b0 = modelo.intercept_
b1 = modelo.coef_[0]
print(f"La ecuación del modelo es: Ventas = {b0:.2f} + {b1:.2f} * Inversión")
```

```python
# 5. Predecir las ventas para una inversión de S/ 8000
inversion_a_predecir = np.array([[8000]])
ventas_predichas = modelo.predict(inversion_a_predecir)

print(f"\nPara una inversión de S/ {inversion_a_predecir[0][0]}, las ventas estimadas son: S/ {ventas_predichas[0]:.2f}")
```

```python
# --- Visualización Gráfica (Opcional) ---
plt.figure(figsize=(10, 6))
# Datos originales
plt.scatter(X, y, color='blue', label='Datos Históricos')
# Línea de regresión
plt.plot(X, modelo.predict(X), color='red', label='Línea de Regresión')
# Punto de la predicción
plt.scatter(inversion_a_predecir, ventas_predichas, color='green', marker='*', s=200, label='Predicción para S/ 8000')
plt.title('Regresión Lineal: Inversión en Publicidad vs. Ventas')
plt.xlabel('Inversión en Publicidad (S/)')
plt.ylabel('Ventas (S/)')
plt.legend()
plt.grid(True)
plt.show()
```
