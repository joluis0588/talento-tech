```python
import pandas as pd

lista_nombres = ["Alexander", "Alejandra", "Dario", "Wilson", "Rosa", "lucia", "juan","carlos", "sofia", "luis", "martha", "alejandro", "lucas", "lucia", "monica", "matias"]
lista_pesos = [69, 57, 64, 80, 40, 62, 70, 68, 50, 71, 66, 79, 67, 52, 70, 48] # peso en kg
lista_alturas = [1.78, 1.6, 1.64, 1.71, 1.53, 1.60, 1.68, 1.57, 1.70, 1.75, 1.67, 1.84, 1.69, 1.58, 1.72, 1.55] # altura en m
# convierte el dicc en un dataframe (marco de datos=dataset=conjunto de datos)
dicc_datos={"nombre":lista_nombres, "peso":lista_pesos,"altura":lista_alturas}
df = pd.DataFrame(dicc_datos)
df
import pandas as pd
import random

# Genera 85 datos adicionales de forma aleatoria para llegar a 100 personas
# (generar 85 datos de personas)
nombres_adicionales = [f'Persona_{i}' for i in range(1, 86)]
pesos_adicionales = [round(random.uniform(45, 120), 2) for _ in range(85)]
alturas_adicionales = [round(random.uniform(1.50, 1.95), 2) for _ in range(85)]

# DataFrame con los datos adicionales
df_adicional = pd.DataFrame({
    'nombre': nombres_adicionales,
    'peso': pesos_adicionales,
    'altura': alturas_adicionales
})

# Concatena los dos DataFrames
df = pd.concat([df, df_adicional], ignore_index=True)

# Calcula el IMC (Índice de Masa Corporal)
# Fórmula: IMC = peso / (altura * altura)
df['imc'] = df['peso'] / (df['altura'] ** 2)

# función para clasificar el peso
def clasificar_peso(imc):
    if imc < 18.5:
        return 'Bajo peso'
    elif 18.5 <= imc < 25:
        return 'Peso normal'
    elif 25 <= imc < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

# columna de clasificación
df['clasificacion'] = df['imc'].apply(clasificar_peso)

# DataFrame final con todas las columnas
print(df)
```
