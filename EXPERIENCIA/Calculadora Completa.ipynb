```python
# Importando la librería/biblioteca de matemáticas de Python:
import math

# DEFINICIÓN DE LAS FUNCIONES:
def mostrar_titulo():
    '''
    input: no hay entradas
    process: mostrar título
    output: no hay salidas
    '''
    print('Ejemplo Calculadora Sencilla (+,-,*,/) por jose luis cuaspud:')

def mostrar_menu():
    '''
    input: no recibe nada (0 entradas)
    process: mostrar menú
    output: no retorna nada (0 salidas)
    '''
    print()
    print("MENU:")
    print("1: sumar (+)")
    print("2: restar (-)")
    print("3: multiplicar (*)")
    print("4: dividir (/)")
    print("5: potenciacion (**)")
    print("6: radicacion (sqrt())")
    print("7: Logaritmo (log10)")
    print("8: seno (sin())")
    print("9: Coseno (cos)")
    print("10: tangente (tan())")
    print("11: salir")

def ingresar_datos():
    '''
    input: no hay entradas
    process: ingresar datos (recibir/leer del usuario)
    output: tres variables que contienen las decisiones del usuario (opcion, numero_uno, numero_dos)
    '''
    opcion = input("Ingrese la opción deseada del menú: ")
    numero_uno = float(input('Ingrese el primer número de la operación:'))
    numero_dos = float(input('Ingrese el segundo número de la operación:'))
    return opcion, numero_uno, numero_dos

def calcular_suma(numero_uno, numero_dos):
    '''
    input: 2 numeros
    process: calcular suma
    output: resultado de la suma
    '''
    resultado = numero_uno + numero_dos
    return resultado

def calcular_resta(numero_uno, numero_dos):
    return numero_uno - numero_dos

def calcular_multiplicacion(numero_uno, numero_dos):
    return numero_uno * numero_dos

def calcular_division(numero_uno, numero_dos):
    if numero_dos != 0:
        return numero_uno / numero_dos
    else:
        return "¡No es posible dividir entre cero!"

def calcular_potenciacion(numero_uno, numero_dos):
    return numero_uno ** numero_dos

def calcular_radicacion(numero_uno, numero_dos):
    return numero_uno ** (1/numero_dos)

def calcular_logaritmo_base10(numero):
    if numero > 0:
        return math.log10(numero)
    else:
        return "Error: logaritmo de un número no positivo."

def calcular_seno(angulo_radianes):
    return math.sin(angulo_radianes)

def calcular_coseno(angulo_radianes):
    return math.cos(angulo_radianes)

def calcular_tangente(angulo_radianes):
    return math.tan(angulo_radianes)

def procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos):
    '''
    input: opcion, numero_uno, numero_dos
    process: procesar y mostrar operacion
    output: no retorna valor
    '''
    print()
    if opcion == '1' or opcion == '+':
        print("Vamos a sumar(+) los dos valores ingresados:")
        resultado = calcular_suma(numero_uno, numero_dos)
        print('La suma es:', resultado)
    elif opcion == '2' or opcion == '-':
        print("Vamos a restar(-) los dos valores ingresados:")
        resultado = calcular_resta(numero_uno, numero_dos)
        print("Resultado =", resultado)
    elif opcion == '3' or opcion == '*':
        print("Vamos a multiplicar(*) los dos valores ingresados:")
        resultado = calcular_multiplicacion(numero_uno, numero_dos)
        print("Resultado =", resultado)
    elif opcion == '4' or opcion == '/':
        print("Vamos a dividir(/) los dos valores ingresados:")
        resultado = calcular_division(numero_uno, numero_dos)
        print("Resultado =", resultado)
    elif opcion == '5' or opcion == '**':
        print("Vamos a calcular la potencia (**) de los dos valores ingresados:")
        resultado = calcular_potenciacion(numero_uno, numero_dos)
        print("Resultado =", resultado)
    elif opcion == '6' or opcion == 'sqrt':
        print("Vamos a calcular la radicación (sqrt) de los dos valores ingresados:")
        resultado = calcular_radicacion(numero_uno, numero_dos)
        print("Resultado =", resultado)
    elif opcion == '7' or opcion == 'log10':
        print("Vamos a calcular el logaritmo (log10) de los dos valores ingresados:")
        resultado1 = calcular_logaritmo_base10(numero_uno)
        resultado2 = calcular_logaritmo_base10(numero_dos)
        print("Logaritmo del primer número:", resultado1)
        print("Logaritmo del segundo número:", resultado2)
    elif opcion == '8' or opcion == 'sin':
        print("Vamos a calcular el seno (sin) de los dos valores ingresados:")
        resultado1 = calcular_seno(numero_uno)
        resultado2 = calcular_seno(numero_dos)
        print("Seno del primer número:", resultado1)
        print("Seno del segundo número:", resultado2)
    elif opcion == '9' or opcion == 'cos':
        print("Vamos a calcular el coseno (cos) de los dos valores ingresados:")
        resultado1 = calcular_coseno(numero_uno)
        resultado2 = calcular_coseno(numero_dos)
        print("Coseno del primer número:", resultado1)
        print("Coseno del segundo número:", resultado2)
    elif opcion == '10' or opcion == 'tan':
        print("Vamos a calcular la tangente (tan) de los dos valores ingresados:")
        resultado1 = calcular_tangente(numero_uno)
        resultado2 = calcular_tangente(numero_dos)
        print("Tangente del primer número:", resultado1)
        print("Tangente del segundo número:", resultado2)
    elif opcion == '11' or opcion == 'salir':
        print("Saliendo de la calculadora. ¡Adiós!")
    else:
        print("Opción no válida. Por favor, intente de nuevo.")

# La función principal que orquesta todo
def llamar_funciones():
    """
    input: no hay entradas
    process: llamar funciones
    output: no hay salidas
    """
    mostrar_titulo()
    mostrar_menu()
    opcion, numero_uno, numero_dos = ingresar_datos()
    procesar_y_mostrar_operacion(opcion, numero_uno, numero_dos)

# EJECUCIÓN DEL PROGRAMA
llamar_funciones()
```
